## Двумерные массивы и форма (shape)

```python
# нули: 3 строки, 4 столбца
a = np.zeros((3, 4))
print(a)
print("shape:", a.shape)
```

```python
# единицы: 2 строки, 5 столбцов
b = np.ones((2, 5))
print(b)
print("shape:", b.shape)
```

```python
# 12 чисел в одну строку
c = np.arange(12)
print(c)
print("shape:", c.shape)
```

```python
c.reshape(3, 4)
```

```python
c = np.arange(1000_000_000)
```

```python
c.reshape(100, 10, 10, 10, 10_00)
```

```python
# те же 12 чисел как 3 строки, 4 столбца
d = np.arange(12).reshape(3, 4)
print(d)
print("shape:", d.shape)
```

```python
# те же 12 чисел как 2 строки, 6 столбцов
e = np.arange(12).reshape(2, 6)
print(e)
print("shape:", e.shape)
```

## Дети выживали чаще взрослых?

```python
# делим возраст на группы: до 18 — ребёнок, иначе взрослый
df['age_group'] = np.where(df['age'] < 18, 'ребёнок', 'взрослый')
df['age_group'].value_counts()
```

```python
# доля выживших в каждой возрастной группе
df.groupby('age_group')['survived'].mean()
```

## «Женщины и дети — вперёд» — правда ли это?

```python
def category(st):
    if st['age'] < 18:
        return 'ребёнок'
    elif st['sex'] == 'female':
        return 'женщина'
    else:
        return 'мужчина'

# применяем к каждой строке
df['category'] = df.apply(category, axis=1)
df['category'].value_counts()
```

```python
# доля выживших по категориям, от большей к меньшей
df.groupby('category')['survived'].mean().sort_values(ascending=False)
```
