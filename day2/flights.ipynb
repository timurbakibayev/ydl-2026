### Что значит одна строка?
Одна строка — это **один месяц одного года**: год, месяц и число пассажиров (в тысячах) на авиарейсах за этот месяц. Например, строка `1949 / Jan / 112` означает: в январе 1949 года перевезли 112 тысяч пассажиров. Всего 12 месяцев × 12 лет = 144 строки.

## Сколько всего пассажиров за все годы?

```python
# суммируем столбец passengers (в тысячах)
flights['passengers'].sum()
```
