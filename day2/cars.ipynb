## 100 Toyotas - синтетические данные

```python
# создаём 100 синтетических Toyotas
n = 100
models = ["Camry", "Corolla", "RAV4", "LC", "Prado", "Highlander", "Yaris", "Avensis"]
cities = ["Almaty", "Astana", "Shymkent", "Karaganda", "Atyrau", "Actau"]

np.random.seed(42)

# базовая цена по модели
base_price = {
    "Camry": 12_000_000,
    "Corolla": 8_000_000,
    "RAV4": 14_000_000,
    "LC": 25_000_000,
    "Prado": 18_000_000,
    "Highlander": 16_000_000,
    "Yaris": 6_000_000,
    "Avensis": 10_000_000,
}

# генерируем модели и годы
model_list = np.random.choice(models, n)
year_list = np.random.randint(2005, 2026, n)

# цена = базовая цена - (2025 - год) * 500_000 + случайный шум
price_list = []
for model, year in zip(model_list, year_list):
    base = base_price[model]
    depreciation = (2025 - year) * 500_000  # теряет ~500k в год
    noise = np.random.randint(-2_000_000, 3_000_000)  # ±2-3 млн случайно
    price = max(2_000_000, base - depreciation + noise)  # минимум 2 млн
    price_list.append(price)

toyotas = pd.DataFrame({
    "make": ["Toyota"] * n,
    "model": model_list,
    "year": year_list,
    "mileage": np.random.randint(0, 300_000, n),
    "price": price_list,
    "city": np.random.choice(cities, n),
})
toyotas.head(10)
```
