```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("911.csv")
```

```python
df.describe()
```

```python
df.info()
```

```python
def category(descr):
    cat = descr.split(":")[0]
    return cat

df["reason"] = df["title"].apply(category)
```

```python
df
```

## Превращаем timeStamp в дату/время и достаём часть, день недели, месяц, дату

```python
# превращаем строку в настоящую дату-время
df["timeStamp"] = pd.to_datetime(df["timeStamp"])

df["timeStamp"]
```

```python
# достаём из даты-времени отдельные части
df["hour"] = df["timeStamp"].dt.hour
df["day_of_week"] = df["timeStamp"].dt.day_name()
df["month"] = df["timeStamp"].dt.month
df["date"] = df["timeStamp"].dt.date

df[["timeStamp", "hour", "day_of_week", "month", "date"]].head()
```

## Какая причина вызова самая частая — EMS, Fire или Traffic?

```python
# считаем, сколько вызовов каждой причины
reason_counts = df["reason"].value_counts()

# строим столбчатый график
reason_counts.plot(kind="bar")
plt.title("Чаще всего звонят по EMS (медицина)")
plt.xlabel("Причина вызова")
plt.ylabel("Количество вызовов")
plt.show()
```

## Сколько вызовов в каждом месяце за весь период

```python
# группируем по месяцу (год-месяц) и считаем вызовы
calls_per_month = df.groupby(df["timeStamp"].dt.to_period("M")).size()

# строим линию
calls_per_month.plot(kind="line", marker="o")
plt.title("Число вызовов падает к концу периода")
plt.xlabel("Месяц")
plt.ylabel("Количество вызовов")
plt.show()
```

## Сколько вызовов по часам суток — когда пик?

```python
# считаем вызовы по каждому часу суток
calls_per_hour = df["hour"].value_counts().sort_index()

# строим столбчатый график
calls_per_hour.plot(kind="bar")
plt.title("Пик вызовов — днём, около 16-17 часов")
plt.xlabel("Час суток")
plt.ylabel("Количество вызовов")
plt.show()
```

## Тот же график по часам, но с разбивкой по причине вызова

```python
import seaborn as sns

# те же часы, но цвет (hue) — причина вызова
sns.countplot(data=df, x="hour", hue="reason")
plt.title("Вызовы по часам с разбивкой по причине")
plt.xlabel("Час суток")
plt.ylabel("Количество вызовов")
plt.show()
```

## Число вызовов по дням недели (с понедельника по воскресенье)

```python
# порядок дней недели
days_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# считаем вызовы по дням и расставляем в нужном порядке
calls_per_day = df["day_of_week"].value_counts().reindex(days_order)

calls_per_day.plot(kind="bar")
plt.title("Вызовы по дням недели")
plt.xlabel("День недели")
plt.ylabel("Количество вызовов")
plt.ylim((11000,16000))
plt.show()
```

## Способ 1: подписать точные числа прямо на столбцах

```python
# столбцы от нуля, но с числом на каждом — глаз видит разницу через цифры
ax = calls_per_day.plot(kind="bar")
ax.bar_label(ax.containers[0])
plt.title("Вызовы по дням недели (с числами на столбцах)")
plt.xlabel("День недели")
plt.ylabel("Количество вызовов")
plt.show()
```

## Способ 2: отклонение от среднего (выше/ниже нуля)

```python
# вычитаем среднее: ноль = средний день, видно кто выше, кто ниже
diff_from_mean = calls_per_day - calls_per_day.mean()

diff_from_mean.plot(kind="bar")
plt.axhline(0, color="black")  # линия среднего
plt.title("Отклонение от среднего по дням недели")
plt.xlabel("День недели")
plt.ylabel("Вызовы минус среднее")
plt.show()
```

## Линейная регрессия: координаты вызовов (lng → lat)
Каждый вызов — точка на карте. Прямая показывает общий наклон облака,
а далёкие точки — это вызовы с явно «битыми» координатами (выбросы).

```python
# scatter + прямая регрессии: x = долгота, y = широта
# scatter_kws — делаем точки мелкими и полупрозрачными, чтобы видеть плотность
sns.regplot(data=df, x="lng", y="lat",
            scatter_kws={"s": 5, "alpha": 0.2},
            line_kws={"color": "red"})
plt.title("Координаты вызовов с линией регрессии и выбросами")
plt.xlabel("Долгота (lng)")
plt.ylabel("Широта (lat)")
plt.show()
```

## Тот же барчарт по дням недели: линия среднего и разница в числах

```python
# среднее число вызовов за день
mean_calls = calls_per_day.mean()

# обычные столбцы от нуля
ax = calls_per_day.plot(kind="bar")

# горизонтальная линия среднего
ax.axhline(mean_calls, color="red")

# на каждом столбце пишем разницу со средним (+/-)
diff = calls_per_day - mean_calls
ax.bar_label(ax.containers[0], labels=[f"{d:+.0f}" for d in diff])

plt.title(f"Вызовы по дням недели (среднее = {mean_calls:.0f})")
plt.xlabel("День недели")
plt.ylabel("Количество вызовов")
plt.show()
```

## Топ-10 конкретных типов вызовов (полный title)

```python
# 10 самых частых типов вызовов
top10 = df["title"].value_counts().head(10)

# горизонтальные столбцы; самый частый — сверху
top10.sort_values().plot(kind="barh")
plt.title("Топ-10 типов вызовов")
plt.xlabel("Количество вызовов")
plt.ylabel("Тип вызова")
plt.show()
```

## Топ-10 районов (township) по числу вызовов

```python
# 10 районов с наибольшим числом вызовов
top10_twp = df["twp"].value_counts().head(10)

# горизонтальные столбцы; самый частый — сверху
top10_twp.sort_values().plot(kind="barh")
plt.title("Топ-10 районов по числу вызовов")
plt.xlabel("Количество вызовов")
plt.ylabel("Район (township)")
plt.show()
```

## Тепловая карта: день недели × час суток

```python
# таблица: строки — дни недели, столбцы — часы, значения — число вызовов

#  EMS, Fire или Traffic
df1 = df[df["reason"]=="Traffic"]

heat = df1.pivot_table(index="day_of_week", columns="hour", values="e", aggfunc="sum")

# расставляем дни в порядке пн -> вс
heat = heat.reindex(days_order)

plt.figure(figsize=(12, 5))
sns.heatmap(heat, cmap="viridis")
plt.title("Вызовы по дню недели и часу суток")
plt.xlabel("Час суток")
plt.ylabel("День недели")
plt.show()
```

## Вызовы по часам: отдельная линия на каждую причину

```python
# таблица: строки — час, столбцы — причина, значения — число вызовов
hour_reason = df.pivot_table(index="hour", columns="reason", values="e", aggfunc="sum")

# каждая колонка-причина становится отдельной линией
hour_reason.plot(kind="line", marker="o")
plt.title("Вызовы по часам суток для каждой причины")
plt.xlabel("Час суток")
plt.ylabel("Количество вызовов")
plt.show()
```

## Карта вызовов: точка на каждый вызов (lng × lat), цвет — причина

```python
# каждая точка — вызов; цвет (hue) — причина
df1 = df[df["lng"] > -76]
df1 = df1[df1["lat"] > 38]
df1 = df1[df1["lat"] < 40.5]
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df1, x="lng", y="lat", hue="reason", s=5, alpha=0.3)
plt.title("Все вызовы на карте округа")
plt.xlabel("Долгота (lng)")
plt.ylabel("Широта (lat)")
plt.show()
```

## Обводим зону зума на полной карте, потом приближаем её через xlim/ylim

```python
import matplotlib.patches as patches

# границы зоны, которую хотим приблизить
x_min, x_max = -76, -75
y_min, y_max = 39.8, 40.5

# полная карта со всеми точками (включая выбросы)
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df, x="lng", y="lat", hue="reason", s=5, alpha=0.3)

# рисуем прямоугольник-рамку вокруг зоны зума
rect = patches.Rectangle((x_min, y_min), x_max - x_min, y_max - y_min,
                         fill=False, edgecolor="red", linewidth=2)
plt.gca().add_patch(rect)

plt.title("Полная карта: красным обведена зона зума")
plt.xlabel("Долгота (lng)")
plt.ylabel("Широта (lat)")
plt.show()
```

```python
# те же данные, но приближаем — просто ограничиваем диапазон осей
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df, x="lng", y="lat", hue="reason", s=5, alpha=0.3)

plt.xlim(x_min, x_max)  # окно по долготе
plt.ylim(y_min, y_max)  # окно по широте

plt.title("Приближение зоны зума (данные те же, обрезаны только оси)")
plt.xlabel("Долгота (lng)")
plt.ylabel("Широта (lat)")
plt.show()
```

## Эффект «лупы»: полная карта и приближённая вставка на одном рисунке

```python
fig, ax = plt.subplots(figsize=(11, 8))

# полная карта со всеми точками
sns.scatterplot(data=df, x="lng", y="lat", hue="reason", s=5, alpha=0.3, ax=ax)
ax.set_title("Полная карта с приближённой вставкой")
ax.set_xlabel("Долгота (lng)")
ax.set_ylabel("Широта (lat)")

# вставка-лупа в углу (координаты вставки в долях рисунка)
axins = ax.inset_axes([0.6, 0.6, 0.38, 0.38])
sns.scatterplot(data=df, x="lng", y="lat", hue="reason", s=5, alpha=0.3,
                ax=axins, legend=False)

# показываем во вставке только нужную зону
axins.set_xlim(x_min, x_max)
axins.set_ylim(y_min, y_max)
axins.set_xlabel("")
axins.set_ylabel("")

# рисует рамку на полной карте и линии-связки к вставке
ax.indicate_inset_zoom(axins, edgecolor="red")

plt.show()
```

## Самый загруженный день: находим его

```python
# число вызовов в каждый день, берём день с максимумом
calls_per_date = df["date"].value_counts()
busiest_day = calls_per_date.idxmax()

print("Самый загруженный день:", busiest_day, "—", calls_per_date.max(), "вызовов")
```

## Вызовы этого дня по часам

```python
# берём только этот день
busy = df[df["date"] == busiest_day]

# вызовы по часам в этот день
busy["hour"].value_counts().sort_index().plot(kind="bar")
plt.title(f"Вызовы по часам — {busiest_day}")
plt.xlabel("Час суток")
plt.ylabel("Количество вызовов")
plt.show()
```

## Какие типы вызовов всплыли в этот день — что случилось?

```python
# топ-10 типов вызовов именно в этот день
top_busy = busy["title"].value_counts().head(10)

top_busy.sort_values().plot(kind="barh")
plt.title(f"Топ типов вызовов — {busiest_day}")
plt.xlabel("Количество вызовов")
plt.ylabel("Тип вызова")
plt.show()
```

## Доказательство: вызовы «DISABLED VEHICLE» по дням за весь период
Если это был снежный шторм, то 23 января должно стоять одиноким пиком над ровным фоном.

```python
# берём только вызовы про застрявшие машины
disabled = df[df["title"] == "Traffic: DISABLED VEHICLE -"]

# считаем их по дням и строим линию во времени
disabled_per_day = disabled.groupby("date").size()

disabled_per_day.plot(kind="line")
plt.title("Вызовы 'DISABLED VEHICLE' по дням — пик 23 января (метель Jonas)")
plt.xlabel("Дата")
plt.ylabel("Количество вызовов")
plt.show()
```

## Отмечаем пиковый час стрелкой прямо на графике

```python
# находим пиковый час и его значение
peak_hour = calls_per_hour.idxmax()
peak_value = calls_per_hour.max()

calls_per_hour.plot(kind="line", marker="o")

# стрелка с подписью к пиковой точке
plt.annotate(f"Пик: {peak_hour}:00 ({peak_value} вызовов)",
             xy=(peak_hour, peak_value),              # куда указывает стрелка
             xytext=(peak_hour + 3, peak_value - 1000),  # где текст
             arrowprops={"arrowstyle": "->", "color": "red"})

plt.title("Вызовы по часам суток с отметкой пика")
plt.xlabel("Час суток")
plt.ylabel("Количество вызовов")
plt.show()
```

## Дни недели по причинам: выделяем Traffic цветом, остальное — серым

```python
# таблица: строки — дни недели, столбцы — причина
day_reason = df.pivot_table(index="day_of_week", columns="reason", values="e", aggfunc="sum")
day_reason = day_reason.reindex(days_order)

# Traffic — синий, остальные причины — серые
colors = ["gray" if reason != "Traffic" else "tab:blue" for reason in day_reason.columns]

day_reason.plot(kind="bar", color=colors)
plt.title("Вызовы по дням недели: Traffic выделен, остальное серым")
plt.xlabel("День недели")
plt.ylabel("Количество вызовов")
plt.show()
```

# Визуальная история: как один снежный шторм перевернул месяц
Пять графиков, каждый — один шаг к выводу. Заголовок графика = его мысль.

```python
# Шаг 1: из чего вообще состоят вызовы — больше всего медицины (EMS)
df["reason"].value_counts().plot(kind="bar")
plt.title("Шаг 1: Обычно звонят по медицине — EMS лидирует")
plt.xlabel("Причина вызова")
plt.ylabel("Количество вызовов")
plt.show()
```

```python
# Шаг 2: у округа есть «норма» — сколько вызовов в день в среднем
calls_per_date_sorted = df.groupby("date").size()
calls_per_date_sorted.plot(kind="line")
plt.title("Шаг 2: У округа есть привычный ритм — около 300-400 вызовов в день")
plt.xlabel("Дата")
plt.ylabel("Вызовов в день")
plt.show()
```

```python
# Шаг 3: тот же ритм, но с отметкой — один день резко выбился
peak_date = calls_per_date_sorted.idxmax()
peak_calls = calls_per_date_sorted.max()

calls_per_date_sorted.plot(kind="line")
plt.annotate(f"{peak_date}\n{peak_calls} вызовов",
             xy=(peak_date, peak_calls),
             xytext=(peak_date, peak_calls + 150),
             arrowprops={"arrowstyle": "->", "color": "red"})
plt.title("Шаг 3: Один день резко выбился из нормы")
plt.xlabel("Дата")
plt.ylabel("Вызовов в день")
plt.show()
```

```python
# Шаг 4: что это был за день — почти все вызовы про застрявшие машины
busy_day = df[df["date"] == peak_date]
top_busy_day = busy_day["title"].value_counts().head(8)

# выделяем DISABLED VEHICLE цветом, остальное серым
bar_colors = ["tab:blue" if "DISABLED VEHICLE" in t else "gray" for t in top_busy_day.index]
top_busy_day.sort_values().plot(kind="barh",
    color=[bar_colors[list(top_busy_day.index).index(t)] for t in top_busy_day.sort_values().index])
plt.title("Шаг 4: В этот день взлетели вызовы о застрявших машинах")
plt.xlabel("Количество вызовов")
plt.ylabel("Тип вызова")
plt.show()
```

```python
# Шаг 5 (вывод): за весь период такое было ровно один раз — снежный шторм
disabled_daily = df[df["title"] == "Traffic: DISABLED VEHICLE -"].groupby("date").size()
spike_date = disabled_daily.idxmax()
spike_value = disabled_daily.max()

disabled_daily.plot(kind="line")
plt.annotate(f"Метель Jonas\n{spike_date}: {spike_value} вызовов",
             xy=(spike_date, spike_value),
             xytext=(spike_date, spike_value - 200),
             arrowprops={"arrowstyle": "->", "color": "red"})
plt.title("ВЫВОД: Один снежный шторм за день дал вызовов больше, чем месяцы вместе")
plt.xlabel("Дата")
plt.ylabel("Вызовов 'DISABLED VEHICLE' в день")
plt.show()
```
