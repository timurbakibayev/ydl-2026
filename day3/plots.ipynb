```python
x = np.linspace(0, 2*np.pi, 200)
plt.plot(x, np.sin(x)+0.3, label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)")
plt.plot(x, np.sin(x) + np.cos(x), label="sin+cos(x)")
plt.legend()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,4))
ax1.plot(x, np.sin(x))
ax1.set_title("sin")

ax2.plot(x, np.cos(x))
ax2.set_title("cos")
plt.savefig("sin_and_cos.png", dpi=300)
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X,Y,Z, cmap="viridis")
plt.show()
```

## Вращающаяся 3D-поверхность (анимация)

```python
%matplotlib widget
from matplotlib import animation

fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(111, projection="3d")

# предел при повороте — диагональ квадрата (5 * sqrt(2))
lim = 5 * np.sqrt(2)

# на каждом кадре пересчитываем X, Y — поворот точек в плоскости на угол angle
def rotate(angle):
    ax.clear()
    t = np.radians(angle)
    Xr = X*np.cos(t) - Y*np.sin(t)   # новые координаты X
    Yr = X*np.sin(t) + Y*np.cos(t)   # новые координаты Y
    ax.plot_surface(Xr, Yr, Z, cmap="viridis")
    # фиксируем масштаб, чтобы кадры не приближались/отдалялись
    ax.set_xlim(-lim, lim)
    ax.set_ylim(-lim, lim)

anim = animation.FuncAnimation(fig, rotate, frames=range(0, 360, 4), interval=50)
plt.show()
```
