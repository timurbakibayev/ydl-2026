```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
x = np.linspace(np.pi*1.5, 2.5*np.pi, 50)
y = np.sin(x)
df = pd.DataFrame({
    "x": x,
    "y": y,
})

X = df.drop("y", axis=1)
y = df["y"]

m = LinearRegression()
m.fit(X, y)
plt.plot(x,y)
plt.plot(x,m.predict(X))
print(m.coef_, m.intercept_)
```

```python
x = np.linspace(0.5*np.pi, 2.5*np.pi, 50)
y = np.sin(x)
df = pd.DataFrame({
    "x": x,
    "y": y,
})

# feature engineering
df["cheating"] = y/2


X = df.drop("y", axis=1)
y = df["y"]

m = LinearRegression()
m.fit(X, y)
plt.plot(x,y)
plt.plot(x,m.predict(X), '-.')
print(m.coef_, m.intercept_)
```

```python
x = np.linspace(0.5*np.pi, 5*np.pi, 50)
y = np.sin(x/2)+np.random.rand(50)/7  # + noice
df = pd.DataFrame({
    "x": x,
    "y": y,
})

# feature engineering
df["sinX"] = np.sin(df["x"])
df["cosX"] = np.cos(df["x"])
df["x2"] = df["x"] ** 2
df["x3"] = df["x"] ** 3


X = df.drop("y", axis=1)
y = df["y"]

m = LinearRegression()
m.fit(X, y)
plt.plot(x,y)
plt.plot(x,m.predict(X), '-.')
print(m.coef_, m.intercept_)
```
