## Назад к простой линии: одна переменная total_bill

```python
# берём свежий датасет и одну переменную
tips = sns.load_dataset("tips")

m = LinearRegression()
X = tips[["total_bill"]]
y = tips["tip"]
m.fit(X, y)

# R^2 — насколько хорошо линия описывает данные
print("R^2:", m.score(X, y))
```

```python
# точки данных и линия модели
tips["predict"] = m.predict(X)

plt.scatter(tips["total_bill"], tips["tip"], label="данные")
plt.plot(tips["total_bill"], tips["predict"], color="red", label="линия модели")
plt.xlabel("total_bill")
plt.ylabel("tip")
plt.title("Чаевые от суммы счёта")
plt.legend()
plt.show()
```
